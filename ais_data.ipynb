```python
import pandas as pd

# data of 5, 6 & 7 December 2023
csv_file_path1 = 'ais1.csv'
columns_to_load1 = ['MMSI', 'BaseDateTime','LAT','LON','SOG','COG']
df1 = pd.read_csv(csv_file_path1, usecols=columns_to_load1)

csv_file_path2 = 'ais2.csv'
df2 = pd.read_csv(csv_file_path2, usecols=columns_to_load1)

csv_file_path3 = 'ais3.csv'
df3 = pd.read_csv(csv_file_path3, usecols=columns_to_load1)


df_combined = pd.concat([df1, df2,df3], ignore_index=True)

print(df_combined)

```

```python
df_combined['BaseDateTime']=pd.to_datetime(df_combined['BaseDateTime'])
df_combined['BaseDateTime']
```

```python
df_combined['day']=pd.DatetimeIndex(df_combined['BaseDateTime']).day
df_combined['time']=pd.DatetimeIndex(df_combined['BaseDateTime']).time
```
